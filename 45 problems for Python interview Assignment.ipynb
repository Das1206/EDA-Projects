```python
from collections import deque  
def postfix_evaluate(items):
    stack = deque()

    for x in items:
        if x=='+' or x=='-' or x=='*' or x=='/':
            op1=stack.pop()
            op2=stack.pop()
            result=calculate(op2,op1,x)
            stack.append(result)
        else:
            stack.append(x)
           
    return stack.pop()

def calculate(op2,op1,x):
    if x == '*':
        return int(op2)*int(op1)
    elif x == '/':
        if (op1 == 0):
            return 0
        else:
            return int(op2)//int(op1)
    elif x == '+':
        return int(op2)+int(op1)
    elif x == '-':
        return int(op2)-int(op1)

items = [2,3,'+',4,'*']
print(postfix_evaluate(items))
items = [2,3,4,'*','+']
print(postfix_evaluate(items))
items = [3,3,3,'-','/']
print(postfix_evaluate(items))
```

```python
def counting_series(n):
    d = 9
    d_sum = 0
    l = 1
    while n - d * l > 0:
        n -= d * l
        d_sum += d
        d *= 10
        l += 1

    r = n % l
    t = n // l
    t += d_sum
    current = t + 1
    out = str(current)[r] 

    return int(out)
```

```python
def spread_the_coins(pile,left,right):
    start=i=0#initializing start and loop is variable
    while i <(len(pile)):
        k = pile[i]//(left+right)#getting k
        if k>0:#the pile being unstable if k>0
            pile[i]-=k*(left+right)#removing extra coins from unstable pile
            if i!=len(pile)-1:
                pile[i+1] += k*right#spreading to right (right exists)
            else:
                pile.append(k*right)#spreading to right by appending
            if i!=0:
                pile[i-1] += k*left#spreading to left (left exists)
                i-=1
                continue#continuing from left spread
            else:
                start-=1#start decrements
                pile.insert(0,k*left)#spreading to right by prepending
                continue#continuing from left spread
        i+=1#advancing
    return(start,pile)
```

```python
from typing import Iterable

def calculate_balance(fulcrum_index, weights):
    return sum(weights[i]*(fulcrum_index-i) for i in range(len(weights)))

def can_balance(weights: Iterable) -> int:
    for i in range(len(weights)):
        if calculate_balance(i, weights) == 0:
            return i
    return -1


if __name__ == '__main__':
    print("Example:")
    print(can_balance([6, 1, 10, 5, 4]))

    # These "asserts" are used for self-checking and not for an auto-testing
    assert can_balance([6, 1, 10, 5, 4]) == 2
    assert can_balance([10, 3, 3, 2, 1]) == 1
    assert can_balance([7, 3, 4, 2, 9, 7, 4]) == -1
    assert can_balance([42]) == 0
    print("Coding complete? Click 'Check' to earn cool rewards!")
```

```python
# Python program to check if rectangles overlap
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

# Returns true if two rectangles(l1, r1)
# and (l2, r2) overlap
def doOverlap(l1, r1, l2, r2):
     
    # To check if either rectangle is actually a line
      # For example  :  l1 ={-1,0}  r1={1,1}  l2={0,-1}  r2={0,1}
       
    if (l1.x == r1.x or l1.y == r1.y or l2.x == r2.x or l2.y == r2.y):
        # the line cannot have positive overlap
        return False
       
     
    # If one rectangle is on left side of other
    if(l1.x >= r2.x or l2.x >= r1.x):
        return False
 
    # If one rectangle is above other
    if(r1.y >= l2.y or r2.y >= l1.y):
        return False
 
    return True
 
# Driver Code
if __name__ == "__main__":
    l1 = Point(0, 10)
    r1 = Point(10, 0)
    l2 = Point(5, 5)
    r2 = Point(15, 0)
 
    if(doOverlap(l1, r1, l2, r2)):
        print("Rectangles Overlap")
    else:
        print("Rectangles Don't Overlap")
 
```

```python
#Define the function

#nearest smaller.

def nearest_smaller(items):

 

    #Define a list to store

    #the resultant list.

    result = []

 

    #Define a variable to

    #store the index of the

    #element currently being compared.

    curr_ele_index = 0

 

    #Run the loop till all the

    #elements have been added

    #to the resultant list.

    while len(result) < len(items):

       

        #Define the variables to

        #store the index of the

        #nearest smaller elements.

        smallest_ele_left = 'x'

        smallest_ele_right = 'x'

 

        left_index = curr_ele_index - 1

 

        #Run the loop to find the

        #nearest smaller element

        #on the left.

        while left_index >= 0:

 

            #If a smaller element is

            #present on the left then,

            #store its index and break

            #out of the loop.

            if(items[left_index] < items[curr_ele_index]):

                smallest_ele_left = left_index

                break

 

            left_index -= 1

 

        right_index = curr_ele_index + 1

 

        #Run the loop to find the

        #nearest smaller element

        #on the right.

        while right_index < len(items):

 

            #If a smaller element is

            #present on the right then,

            #store its index and break

            #out of the loop.

            if(items[right_index] < items[curr_ele_index]):

                smallest_ele_right = right_index

                break

 

            right_index += 1

 

        #If there was no smaller

        #element present on the left then,

        #check the if there was a

        #smaller element on the right or not.

        if smallest_ele_left == 'x':

 

            #If there is a smaller element

            #on the right then, append it

            #to the resulatant list..

            if smallest_ele_right == 'x':

                result.append(items[curr_ele_index])

 

            #Otherwise, append the current

          #element to the resultant list.

            else:

                result.append(items[smallest_ele_right])

 

        #Otherwise, move to

        #the else case.

        else:

 

            #If there is no smaller element

            #on the right then, append the

            #nearest left smaller element

            #to the resulatant list.

            if smallest_ele_right == 'x':

                result.append(items[smallest_ele_left])

 

            #Otherwise, compare both

            #the smaller elements.

            else:

 

                #Find the distance of

                #both the smaller elements.

                dist_left = curr_ele_index - smallest_ele_left

                dist_right = smallest_ele_right - curr_ele_index

 

                #If the distance of the left

                #element is less than the right

                #element then, append the left

                #element to the resultant list.

                if(dist_left < dist_right):

                    result.append(items[smallest_ele_left])

 

                #Otherwise, if the distance of

                #the right element is smaller

                #than the left element then,

                #append the right element

                #to the resultant list.

                elif (dist_right < dist_left):

                    result.append(items[smallest_ele_right])

 

               

                else:

                    smaller_ele = min(items[smallest_ele_left], items[smallest_ele_right])

                    result.append(smaller_ele)

 

        curr_ele_index += 1

   

    #Return the

    #resultant list.

    return result
```

```python
def reverse_ascending_sublists(items):
    output = []  # type: List[int]
    sublist = []  # type: List[int]
    for item in items:
        if len(sublist) == 0:
            sublist.append(item)
        else:
            if item > sublist[0]:
                sublist.insert(0, item)
            else:
                output.extend(sublist)
                sublist = [item]
    output.extend(sublist)
    return output
```

```python
def collapse_intervals(items):
    start = end = None  
    output = []  
    for item in items:
        if start is None:
            start = end = item
        elif item == end + 1:
            end = item
        else:
            if start == end:
                interval = str(start)
            else:
                interval = str(start) + '-' + str(end)
            output.append(interval)
            start = end = item
    if start == end:
        interval = str(start)
    else:
        interval = str(start) + '-' + str(end)
    output.append(interval)
    return ','.join(output)
```

```python
# count word dominators


def flatten(items):
    result = []
    for x in items:
        if hasattr(x, '__iter__'):  # is x an iterable?
            result.extend(flatten(x))
        else:
            result.append(x)
    return result
```

```python
def collatz(start):
    curr, result = start, []
    if curr < 1:
        return []
    while curr > 1:
        result.append(curr)
        curr = curr // 2 if curr % 2 == 0 else 3 * curr + 1
    result.append(1)
    return result
```

```python
def hangman(wordlist, blank='*'):
    misses, guesses = 0, []
    secret = random.choice(wordlist)
    displayed = [blank for _ in secret]
    print("Let us play a game of Hangman.")
    while blank in displayed:
        print(f"Current word is {'-'.join(displayed)}.")
        print(f"Letters guessed so far: {''.join(guesses)}")
        guess = input("Please enter a letter, or 'quit' to quit: ")
        if guess == 'quit':
            return None
        if guess not in guesses:
            guesses.append(guess)
            guesses.sort()
        hit = 0
        for (i, (c1, c2)) in enumerate(zip(displayed, secret)):
            if c1 == blank and c2 == guess:
                displayed[i] = guess
                hit += 1
        if hit > 0:
            print(f"That one hit {hit} time{'s' if hit > 1 else ''}!")
        else:
            print("That one was a miss.")
            misses += 1
    print(f"You guessed {secret!r} with {misses} misses.\n")
    return misses
```
